## Exploratory Data Analysis (EDA)

### Explore Data Day

##### Jumlah data day sepeda yang dipakai berdasarkan holiday, season, weekday, dan workingday

```python
data = day_df[['holiday', 'season', 'weekday', 'workingday', 'cnt']]

jumlah = data.groupby(['holiday', 'season', 'weekday', 'workingday'])['cnt'].sum().reset_index()

jumlah = jumlah.sort_values(by='cnt', ascending=False)
jumlah
```

##### Jumlah data day sepeda dipakai berdasarkan weathersit, temp, atemp, hum, windspeed

```python
faktor_data = day_df[['weathersit', 'temp', 'atemp', 'hum', 'windspeed', 'cnt']]

jumlah_faktor = faktor_data.groupby(['weathersit', 'temp', 'atemp', 'hum', 'windspeed'])['cnt'].sum().reset_index()

jumlah_faktor = jumlah_faktor.sort_values(by='cnt', ascending=False)

jumlah_faktor
```

## Visualization & Explanatory Analysis

### Pertanyaan 1 : Apakah ada pengaruh cuaca terhadap jumlah peminjam sepeda?

```python

plt.figure(figsize=(10, 6))
ax = sns.barplot(x='weathersit_y', y='cnt_y', data=df, ci=None)

for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x() + p.get_width() / 2., height + 0.1,
            f'{height / len(df) * 100:.1f}%', ha='center')

plt.title('Rata-rata Jumlah Pengguna Sepeda per Jenis Cuaca')
plt.xlabel('Jenis Cuaca')
plt.ylabel('Rata-rata Jumlah Pengguna Sepeda')
plt.show()
```

### Pertanyaan 2: Bagaimana tren penggunaan sepeda berdasarkan jam dalam sehari?

```python

plt.figure(figsize=(14, 6))
sns.pointplot(x='hr', y='cnt_y', data=df, hue='season_y', markers='o', linestyles='-', ci=None)
plt.title('Jumlah Pengguna Sepeda per Jam dalam Sehari')
plt.xlabel('Jam dalam Sehari')
plt.ylabel('Jumlah Pengguna Sepeda')
plt.legend(title='Musim')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

df['dteday'] = pd.to_datetime(df['dteday'])

df_sorted = df.sort_values(by='dteday')

plt.figure(figsize=(14, 8))
sns.lineplot(x='dteday', y='cnt_x', data=df_sorted, ci=None)
plt.xlabel('Tanggal')
plt.ylabel('Jumlah Peminjaman Sepeda')
plt.title('Performa Peminjaman Sepeda dalam Beberapa Bulan Terakhir')
plt.xticks(rotation=0)
plt.show()
```
